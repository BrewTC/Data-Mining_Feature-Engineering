# 0919 thesis

```python
from apyori import apriori
import pandas as pd
```

```python
df = pd.read_csv('0919 Apriori df.csv') # names=['products'], header=None

features = pd.read_csv('0919 Apriori features.csv')
features = list(features['features'])
```

```python
df_str = df[features].astype(str)
# df_str.dtypes

df_dum = pd.get_dummies(df_str)
df_dum.head()
df_dum.to_csv('0920 df_dum.csv', index = None)
```

```python
df_dum_int = df_dum.astype('int')
df_dum_int_bool = df_dum_int.astype('bool')
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# 找出 frequent itemsets(頻繁項集)，最低支持度 (min support) 定為 50%
frequent_itemsets = apriori(df_dum_int_bool, min_support=0.5, use_colnames=True)
frequent_itemsets.to_csv('0920 frequent_itemsets.csv', index = None)
```

```python
help(apriori)
```

```python
association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)
```

```python
a_rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
a_rules.sort_values('lift', ascending = False) # 由最高到最低排序
```

```python
gg = a_rules.sort_values('lift', ascending = False)
gg.to_csv('0920 association_rules result.csv')
```

# 0930 Test

```python
from apyori import apriori
import pandas as pd
```

```python
df = pd.read_csv('0930 second test(add BS).csv')
df.head()
```

```python
df.info()
```

```python
df_dropna = df.dropna()
df_dropna = df_dropna.astype('int')
df_dropna.head(10)
```

```python
df_dropna.info()
```

```python
df_dropna['real_hypertension'].unique()
```

```python
df_dropna_bool = df_dropna.astype('bool')
df_dropna_bool.head()
```

```python
df_dropna_bool.info()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# 找出 frequent itemsets(頻繁項集)，最低支持度 (min support) 定為 60%
frequent_itemsets = apriori(df_dropna_bool, min_support=0.5, use_colnames=True)

frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.6) ]
```

### 從 frequent itemsets 中 generate association rules
### 關聯規則（associational rules）: 暗示兩種物品之間可能存在很強的關係。

```python
association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)
```

```python
a_rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
a_rules.sort_values('lift', ascending = False) # 由最高到最低排序
```

```python
a_rules[      # (a_rules['length'] >= 2) &
              (a_rules['confidence'] > 0.6) &
              (a_rules['lift'] > 1)              ]
```
